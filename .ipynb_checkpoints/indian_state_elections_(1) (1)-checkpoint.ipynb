# Data Preprocessing

```python
# Removing Unncessary Columns
def remove_unncessary_cols(df, cols):
    df = df.drop(cols, axis = 1)
    print(f"Shape of the data after deleteing columns is {df.shape}")
    return df

# Columns to be removed
cols = ['Unnamed: 0', 'Username', 'Tweet Id']

# Removing columns that are unnecessary
goa_df = remove_unncessary_cols(goa_df, cols)
manipur_df = remove_unncessary_cols(manipur_df, cols)
punjabElection_df = remove_unncessary_cols(punjabElection_df, cols)
upElections_df = remove_unncessary_cols(upElections_df, cols)
uk_df = remove_unncessary_cols(uk_df, cols)
```

```python
# Counting duplicates in each data
def count_duplicates(df):
    df = df[df.duplicated()]
    length = len(df)
    print(f'There are {length} duplicates in data')

# Counting duplicates
count_duplicates(goa_df)
count_duplicates(manipur_df)
count_duplicates(punjabElection_df)
count_duplicates(upElections_df)
count_duplicates(uk_df)
```

- Since there are 2 duplicate rows present in UP Election data so we drop the duplicates manually.

```python
# dropping duplicates from UP Election Data
upElections_df = upElections_df.drop_duplicates()
```

```python
# Converting datetime object to data and extract day, month, year, time and add a the election column
def convert_datetime(df, date, electionType):
    # convert datetime to datetime format
    df[date] = pd.to_datetime(df[date], utc = True, errors = 'coerce')

    # extracting the datetime component from the date
    # Extracting date components and add them as new columns
    df['Year'] = df[date].dt.year
    df['Month'] = df[date].dt.month
    df['Day'] = df[date].dt.day
    df['Hour'] = df[date].dt.hour
    df['Minute'] = df[date].dt.minute

    # Adding new 'Party' column with the specified value
    df['ElectionType'] = electionType

    return df
```

```python
# Convert datetime column for each dataframe
goa_df = convert_datetime(goa_df, 'Datetime', 'Goa')
manipur_df = convert_datetime(manipur_df, 'Datetime', 'Manipur')
punjabElection_df = convert_datetime(punjabElection_df, 'Datetime', 'Punjab')
upElections_df = convert_datetime(upElections_df, 'Datetime', 'Uttar Pradesh')
uk_df = convert_datetime(uk_df, 'Datetime', 'Uttarakhand')
```

```python
# Deleting Unnecessary columns
df = df.drop(['Unnamed: 0', 'Datetime'], axis = 1)
```

```python
# Remove Emoji from the text and convert it to the correspoding text
def remove_emoji(text):
    # Convert emojis to their corresponding text descriptions
    text_with_descriptions = emoji.demojize(text, language='en')

    # Remove the colons and replace underscores with spaces
    cleaned_text = re.sub(r':', '', text_with_descriptions).replace('_', ' ')

    return cleaned_text

df['Text'] = df['Text'].apply(remove_emoji)
```

```python
# Take 10000 rows of the data
sampled_df = df.sample(n=1000, random_state=42)
```

```python
# Translate the text in English for further preprocessing
translator = Translator() # intialize translator
def translate_text(text):
    try:
        translation = translator.translate(text, dest='en')
        return translation.text
    except:
        return text

# translate the dext
sampled_df['Translated_Text'] = sampled_df['Text'].apply(translate_text)
```

```python
# sampled df
sampled_df.head()
```

```python
# Function for text preprocessing
def lower_text(text):
    # convert text to lower case
    text = text.lower()
    return text


# Remove stopwords
def remove_stopwords(text):
    stopwords_ = set(stopwords.words('english'))
    text = ' '.join([word for word in text.split() if word not in stopwords_])
    return text

# Remove punctuation
def remove_punctuation(text):
    text = re.sub(r'[^\w\s]', '', text)  # Remove punctuation
    return text

# Remove URLs
def remove_urls(text):
    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)
    return text

# Remove numbers
def remove_numbers(text):
    text = re.sub(r'\d+', '', text)
    return text

# Remove extra whitespaces
def remove_extra_whitespaces(text):
    text = re.sub(r'\s+', ' ', text).strip()  # Remove extra spaces and leading/trailing spaces
    return text

# Combine all preprocessing steps into a single function
def preprocess_text(text):
    text = lower_text(text)
    text = remove_stopwords(text)
    text = remove_punctuation(text)
    text = remove_urls(text)
    text = remove_numbers(text)
    text = remove_extra_whitespaces(text)
    return text

```

```python
# Apply Text preprocessing
sampled_df['Translated_Text'] = sampled_df['Translated_Text'].apply(preprocess_text)
```

```python
# Calculating metrics such as word count, character counts,average word
sampled_df['Word count'] = sampled_df['Translated_Text'].apply(lambda x: len(x.split())) # Calculating word counts
sampled_df['Char counts'] = sampled_df['Translated_Text'].apply(lambda x: len(x)) # Calculating character counts
sampled_df['Average word counts'] = sampled_df['Word count']/sampled_df['Char counts'] # average word counts
```

```python
#reset the index
sampled_df = sampled_df.reset_index()
```

```python
# drop columns
sampled_df = sampled_df.drop(['index'], axis = 1)
```
